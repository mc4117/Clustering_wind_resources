```python
# Determine clusters when clustering wind velocity data using 6 clusters

cluster_num = 6
u_arr_reshape = u_combined.reshape(u_combined.shape[0], u_combined.shape[1]*u_combined.shape[2]*2)
kmeans = KMeans(n_clusters=cluster_num, random_state=0).fit(u_arr_reshape)
labels_df = pd.DataFrame(kmeans.labels_)
cluster_centres = kmeans.cluster_centers_.reshape(cluster_num,u_combined.shape[1], u_combined.shape[2], u_combined.shape[3])

mean_err_list = []
mean_err_cluster_list = []
mean_corr_list = []
u_all = []
all_pattern_labels = []

for i in range(cluster_num):
    pattern_labels = labels_df[labels_df[0] == i].reset_index()['index']
    all_pattern_labels.append(pattern_labels)
    u_pattern_0 = np.array([u_combined[j] for j in pattern_labels])
    u_mean = u_pattern_0.mean(axis = 0)
    u_cluster_centre = cluster_centres[i]

    err_list = []    
    corr_list = []
    u_pattern_list = []
    for l in u_pattern_0:
        u_pattern_list.append(l)
        err_list.append(np.abs(l - u_mean).sum())
        corr_list.append(np.corrcoef(l[:, :, 0].reshape(153*72), u_mean[:, :, 0].reshape(153*72))[0][1] + np.corrcoef(l[:, :, 1].reshape(153*72), u_mean[:, :, 1].reshape(153*72))[0][1])
    u_all.append(u_pattern_list)
    mean_err_list.append((np.abs(err_list)))
    mean_corr_list.append(np.array(corr_list))
        
    err_cluster_list = []    
    for l in u_pattern_0:
        err_cluster_list.append(np.abs(l - u_cluster_centre).sum())
    mean_err_cluster_list.append((np.abs(err_cluster_list)))

labels = kmeans.labels_
print(silhouette_score(u_arr_reshape, labels, metric = 'euclidean'))
```
